```python
# 📂 Load Clean Data

import pandas as pd

df = pd.read_csv("data/cleaned/clean_retail.csv")

df.head()
```

```python
# 👤 Create Customer-Level Summary

customer_df = df.groupby('CustomerID').agg({
    'InvoiceNo': 'nunique',      # Number of orders
    'Quantity': 'sum',           # Total quantity purchased
    'Revenue': 'sum'             # Total revenue generated
}).reset_index()

customer_df.columns = ['CustomerID', 'Total_Orders', 'Total_Quantity', 'Total_Revenue']

customer_df.head()
```

```python
# 💰 Average Order Value

customer_df['Avg_Order_Value'] = customer_df['Total_Revenue'] / customer_df['Total_Orders']
```

```python
# ⏳ Recency (Days since last purchase)

import datetime as dt

latest_date = df['InvoiceDate'].max()

recency_df = df.groupby('CustomerID')['InvoiceDate'].max().reset_index()

recency_df['Recency'] = (latest_date - recency_df['InvoiceDate']).dt.days

recency_df = recency_df[['CustomerID', 'Recency']]

recency_df.head()
```

```python
# 🔗 Merge Customer Data

customer_df = customer_df.merge(recency_df, on='CustomerID')

customer_df.head()
```

```python
# 📊 Customer Insights

customer_df.describe()
```

```python
# 🏆 Top Customers by Revenue

customer_df.sort_values(by='Total_Revenue', ascending=False).head(10)
```

```python
# ⚠️ Low Value Customers

customer_df.sort_values(by='Total_Revenue').head(10)
```

```python
# 💾 Save Customer Dataset

customer_df.to_csv("data/cleaned/customer_data.csv", index=False)
```
